```python
## Retrieval augmented generation

import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['OPENAI_API_KEY']=os.getenv("OPENAI_API_KEY")
```

```python
from llama_index.core.readers import SimpleDirectoryReader
documents=SimpleDirectoryReader("data").load_data()
```

```python
documents
```

```python
from llama_index.core import VectorStoreIndex,GPTVectorStoreIndex
index = VectorStoreIndex.from_documents(documents)
test_index = GPTVectorStoreIndex.from_documents(documents)
```

```python
engine = index.as_query_engine()
test_engine = test_index.as_query_engine()
```

```python
from llama_index.core.retrievers import VectorIndexRetriever
from llama_index.core.postprocessor import SimilarityPostprocessor
from llama_index.core.query_engine import RetrieverQueryEngine

retriver = VectorIndexRetriever(index,similarity_top_k=4)
postprocessor = SimilarityPostprocessor(similarity_cutoff=0.80)

query_engine=RetrieverQueryEngine(retriever=retriver,
                                  node_postprocessors=[postprocessor])
```

```python
response=query_engine.query("what is The Myth of the Genius Programmer")
```
